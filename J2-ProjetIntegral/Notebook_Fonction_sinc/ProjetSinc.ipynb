```python
from ProjectPi import PiLeibniz, PiEuler
from math import pi
import matplotlib.pyplot as plt
n = 50
L1, L2, L3 = [], [], []
for i in range(n):
    ErrLeibniz = abs(pi-PiLeibniz(i))
    ErrEuler = abs(pi-PiEuler(i))
    L1.append(i)
    L2.append(ErrLeibniz)
    L3.append(ErrEuler)

plt.plot(L1, L2, label = "Erreur Leibniz")
plt.plot(L1, L3, label = "Erreur Euler")
plt.legend()
```

### Schéma d'Euler
$$\pi = \sqrt{\sum_{k = 1}^{\infty} \frac{1}{k^2}}$$

```python
import numpy as np
x = np.linspace(-2*np.pi, 2*np.pi, 201)
y = x
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sinc(R)
plt.imshow(Z, cmap = "jet")
plt.colorbar()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
surf = ax.plot_surface(X, Y, Z, cmap = "jet")
fig.colorbar(surf, shrink = 0.5, aspect = 10)
plt.savefig("sinc3D.png")
plt.savefig("sinc3D.pdf")
```
