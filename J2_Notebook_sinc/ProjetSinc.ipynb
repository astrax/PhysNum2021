# Introduction

Nous allons beaucoup travailler avec [Jupyter Notebooks](http://jupyter.org/) (anciennement IPython Notebooks) dans ce cours. Jupyter est une application Web qui regroupe intimement deux fonctionnalités très différentes :

* Un outil qui permet de créer des documents multimédia intégrant du texte, des formules mathématiques, des graphiques, des images, voire des animations et des vidéos.

* Une interface qui permet d'exécuter du code informatique. Pour cela Jupyter s'appuie sur des programmes indépendants capable d'interpréter le langage dans lequel est écrit ce code. Dans la terminologie de Jupyter ces interpréteurs sont appelés des noyaux (kernel en anglais).

Les documents Jupyter sont appelés des notebooks. Un fichier notebook est reconnaissable par son extension .ipynb.

La meilleure façon de comprendre les notebooks est de les essayer, alors commençons!

# Lancez une session de notebook Jupyter sur votre ordinateur (ou smartphone)
Pour lancer le serveur notebook, ouvrez d’abord un terminal, puis entrez:

```
jupyter notebook
```
Cela va démarrer le serveur notebook. Il doit ouvrir automatiquement la page principale du serveur de notebook dans votre navigateur par défaut. Dans le cas contraire, ouvrez simplement le navigateur de votre choix et entrez http://localhost:8888/tree dans la barre d'adresse.

Cela fera apparaître un simple navigateur de fichiers qui affichera le contenu du répertoire à partir duquel vous avez lancé le terminal. Cliquez sur le bouton *New Notebook*, puis sélectionnez *Python 3* pour créer votre premier notebook.

![newnotebook](./media/nb1.gif)

# Exécuter une cellule de code

Au-dessous des barres d’outils, vous verrez une seule cellule de code unique, précédée de `In [:]`. Cette cellule peut contenir un segment de code arbitrairement long, mais nous commencerons par une simple ligne. Dans cette cellule de code, tapez:
```python
x = 2
```
puis appuyez sur **Shift + Entrée**. Si vous appuyez simplement sur **Entrée**, vous constaterez que cela ajoutera simplement une autre ligne à la cellule actuelle. Alors, rappelez-vous, pour exécuter une cellule, il suffit d'appuyer sur **Shift + Entrée**.

Alors, qu'est-ce-qu'il s'est passé? Nous avons assigné l’étiquette x au nombre 2. Et vous devriez également voir que l’étiquette de cette cellule se lira désormais `In [1]:` parce que c’est la première instruction que nous avons exécutée dans ce noyau Python. Vous remarquerez également que le notebook a créé une nouvelle cellule, car nous utilisions déjà la seule cellule existante.

Dans cette nouvelle cellule, essayons d’afficher la valeur attribuée à x. Dans un programme Python la fonction `print()` permet d'écrire sur la sortie standard. entrez donc:

```python
print(x)
```
puis appuyez sur **Shift + Entrée**. Et voici le résultat attendu! La cellule est étiquetée `In [2]:` et le résultat de cette commande est affiché immédiatement sous la cellule.

L'ensemble de la procédure devrait ressembler à ceci:

# Le noyau (kernel)
Ne vous inquiétez pas trop de ce qu'est le noyau (ou kernel en anglais), mais le point principal à retenir ici est que nous pouvons affecter une variable dans une cellule tout en y ayant accès dans une cellule séparée. Les cellules sont des moyens pour diviser nos pensées et notre code, mais tout est connecté en dessous.

# Écrasement des variables
Étant donné que chaque cellule interagit avec la même instance Python, si nous donnons une nouvelle valeur à `x` puis que nous saisissons `print (x)`, nous obtiendrons cette nouvelle valeur. C'est assez simple - mais si nous supprimions ensuite la cellule où nous avons donné une nouvelle valeur à `x`?

Jetons un coup d'oeil!
![runandprint](./media/nb3.gif)

Même si nous avons supprimé la cellule où nous avons affecté `x = 5`, l’affectation est toujours valide. En fait, l’affectation restera valide jusqu’à ce que nous exécutions explicitement une cellule définissant `x` sur une nouvelle valeur ou jusqu’à ce que nous redémarrions complètement cette instance de Jupyter Notebook.

# Markdown
Markdown est un * format d’écriture * qui facilite la saisie de texte bien formaté qui est rendu en XHTML correctement formaté. C'est vraiment génial. Les cellules des blocs-notes Jupyter peuvent être utilisées pour de nombreuses choses: exécuter Python, incorporer des médias ou écrire du texte dans Markdown. Cela nous permet d'écrire des notes sur ce que nous faisons, ce que fait le code, ce que nous *essayons* de faire, comme nous voulons! Ces notes peuvent être pour nous-mêmes, pour documenter notre travail ou pour partager avec d’autres.

Pour créer une cellule Markdown dans un notebook, cliquez sur une cellule vide, puis cliquez sur la liste déroulante (par défaut, placée sur « Code ») et sélectionnez « Markdown » -comme indiqué ci-dessous.

Markdown est également (en quelque sorte) un code. Ainsi, après avoir tapé du texte, vous devez également appuyer sur **Shift + Entrée** pour exécuter la cellule et restituer le texte de markdown. Essaye le! Tapez simplement une phrase ou deux dans une cellule de markdown, puis appuyez sur **Shift + Entrée** pour rendre le texte.
![Markdown](./media/nb4.gif)

# Markdown Math (MathJax)
Markdown peut faire plus que simplement restituer du texte simple. Il est également possible d'insérer des morceaux de texte en langage [*MathJax*](http://docs.mathjax.org/en/latest/mathjax.html) (presque identique au langage *LaTeX*) pour composer des expressions mathématiques.
- L'intégration HTML de formules est possible en les définissant sous forme d'expressions *MathJax* entourées du caractère dollar `$`: `$...$`
- Pour un rendu sur une seule ligne, placez *MathJax* entre deux signes `$$`:
`$$...$$`
![LaTex](./media/nb5.gif)
**Remarque:**
Sachez que la composition mathématique est gérée par *MathJax* et non par *LaTeX*. Bien que la grande majorité de la syntaxe *MathJax* soit identique à *LaTeX*, il existe quelques petites différences (en particulier en ce qui concerne les commandes de matrice). Ainsi, si vous constatez que quelque chose ne se présente pas comme vous le souhaitez, vous pouvez consulter Google pour vérifier que vous utilisez la bonne commande.

## Bibliothèques numpy et matplotlib

### Colormap: Tracés contour, Imshow et 3D


Voir la documentation de matplotlib colormaps <http://matplotlib.org/users/colormaps.html>.

#### Tracés contour

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 201)
y = x
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sinc(R)
## Figure
plt.figure(figsize=(7, 5))
plt.contourf(X, Y, Z, cmap = "jet")
C =plt.contour(X, Y, Z, colors = "black")
plt.clabel(C, inline=1, fontsize=10)
plt.colorbar()
````

#### Imshow (Image pixelisée)

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 50)
y = x
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sinc(R)
## Figure
plt.figure(figsize=(7, 5))
plt.imshow(Z, cmap = "jet")
plt.colorbar()
````

#### Tracé en 3D

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi, np.pi, 50)
y = x
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sinc(R)
## Figure
fig = plt.figure(figsize=(7, 5))
ax = Axes3D(fig)
surf = ax.plot_surface(X, Y, Z, cmap = "jet")
fig.colorbar(surf, shrink = 0.5, aspect = 10)
ax.contourf(X, Y, Z, zdir='z', offset=-1, cmap="jet")
ax.set_zlim(-1, 1)
plt.savefig("sinc3D.png")
plt.savefig("sinc3D.pdf")
````
